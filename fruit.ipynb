```python
from glob import glob

img_list = glob('fruit_dataset/train/images/*.jpg')

print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))
```

```python
with open('fruit_dataset/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('fruit_dataset/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml

with open('fruit_dataset/data.yaml', 'r') as f:
  data = yaml.load(f)

print(data)

data['train'] = 'fruit_dataset/train.txt'
data['val'] = 'fruit_dataset/val.txt'

with open('fruit_dataset/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
